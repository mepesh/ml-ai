```python
#basic plot
plt.plot([1,2,3],[4,1,2])
```

```python
#labelling on both axis
plt.plot([1,2,3],[4,1,2])
plt.xlabel('XAXIS')
plt.ylabel('YAXIS')
plt.title('Basic Graph')
plt.show()
```

```python
#plotting multiple lines
dev_y=[25458,31255,45698,51250,62360]
dev_x=[2,4,6,8,10]

py_dev_y=[38125,48125,55180,66789,90000]

plt.plot(dev_x,dev_y,label="All devs")

plt.plot(dev_x,py_dev_y, label="Python")

plt.xlabel('Year of Experience')
plt.ylabel('Salary')
plt.title('Github salary Chart')

#distingusing graphs
# plt.legend(["All Devs","python only"])
plt.legend()
# style.use("ggplot")
```

```python
#coloring and lines and markers
plt.plot(dev_x,dev_y, color='#142536', marker='.' , label="All Devs")
plt.plot(dev_x,py_dev_y, color='#183645', marker="o", label="Python")
plt.legend()
```

```python
#configure lines
plt.plot(dev_x,dev_y, color='#142536', linewidth=3, label="All Devs")
plt.plot(dev_x,py_dev_y, color='#993645',linewidth=3, label="Python")
plt.legend()
plt.show()
```
